```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
lambda_rate = 0.0005      # Rate parameter λ
scale = 1 / lambda_rate  # Exponential distribution scale = 1/λ i.e. the mean
n_samples = 10000

# 1. Generate random samples
samples = np.random.exponential(scale=scale, size=n_samples)

# 2. Plot histogram of samples (density normalized)
plt.hist(samples, bins=50, density=True)

# 3. Overlay theoretical PDF
x = np.linspace(0, samples.max(), 200)
pdf = lambda_rate * np.exp(-lambda_rate * x)
plt.plot(x, pdf)

# 4. Labels and title
plt.title("Exponential Distribution (λ = 1.5)")
plt.xlabel("x")
plt.ylabel("Density")

# 5. Show plot
plt.show()
```
